```python
# Set the folder path for aggressive and conservative data.
agg_folder = 'Record/Aggressive/Data/'
con_folder = 'Record/Conservative/Data/'

# Initialize lists to hold CSV file paths.
agg_txt_list = []
con_txt_list = []

# Append CSV file paths to the lists.
for file in glob.glob(agg_folder + '*.txt'):
    agg_txt_list.append(file)
for file in glob.glob(con_folder + '*.txt'):
    con_txt_list.append(file)

NUM_ACT = 8
```

```python
agg_results = [optimize_for_two_modules(f, NUM_ACT, False) for f in agg_txt_list]
con_results = [optimize_for_two_modules(f, NUM_ACT, False) for f in con_txt_list]
```

```python
histo_vis(agg_results, False, False, fontsize=25) # logScale = Fasle or True
histo_vis(con_results, False, False, fontsize=25) # logScale = Fasle or True
```
